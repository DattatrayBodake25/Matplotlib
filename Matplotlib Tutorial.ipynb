# BAR PLOT




```python
#Bar plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



x = [1,2,3,4,5]
y = [9,8,7,6,5]

c = ["r","g","y","m","b"]

plt.bar(x,y,color=c,width=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Pyhton","Java","C","C++","R"]

y = [99,96,90,92,60]

c = ["r","g","y","m","b"]
plt.xlabel("Programming Languages",fontsize=16)
plt.ylabel("Grade Numbers",fontsize=16)
plt.title("$CODING$",fontsize=20)
plt.bar(x,y,color=c,width=0.5,align="center",edgecolor="black",linewidth=5,alpha=1,label="Data")
plt.legend()
plt.show()
```
